```python
def kalman_filter_R(column):
    if column=='rotation_rate_x' or column=='rotation_rate_y':
        return 10
    elif column=='rotation_rate_z':
        return 500
    elif column=='linear_accel_x' or column=='linear_accel_y' or column=='linear_accel_z' or column=='throttle_sensor':
        return 100
    elif column=='brake':
        return 1
    elif column=='steering':
        return 10
    elif column=='brake_sensor':
        return 50
    elif column=='steering_sensor':
        return 17
    elif column=='meanEffTorque':
        return 250
    elif column=='FL_wheel_speed' or column=='FR_wheel_speed' or column=='RL_wheel_speed' or column=='RR_wheel_speed':
        return 99
    elif column=='steering_speed' or column=='vehicle_speed':
        return 5
    elif column=='throttle':
        return 0.1
    elif column=='value':
        return 200
```

```python
def apply_kalman_filter(dataframe):
    columns=dataframe.columns.tolist()
    columns.remove('utime')
    filtered_df=pd.DataFrame()
    filtered_df['utime']=dataframe['utime']
    for i in columns:
        attribute_to_filter = i
        measurements = df[attribute_to_filter].values
    
        kf = KalmanFilter(dim_x=2, dim_z=1)

        kf.x = np.array([measurements[0], 0])

        dt = 1 
        kf.F = np.array([[1, dt],
                 [0, 1]])
    
        kf.H = np.array([[1, 0]])

        
        kf.R *= kalman_filter_R(attribute_to_filter)


        filtered_readings = []
        for z in measurements:
            kf.predict()
            kf.update(z)
            filtered_readings.append(kf.x[0])
        filtered_df[i]=filtered_readings
    return filtered_df
```
