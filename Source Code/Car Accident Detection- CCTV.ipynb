```python
from inference_sdk import InferenceHTTPClient
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import requests
from io import BytesIO
```

```python
CLIENT = InferenceHTTPClient(
    api_url="https://detect.roboflow.com",
    api_key="0KvdmSkxwXj7MEVa1vjX"
)
```

```python
image_path=r"C:\Users\mbsas\OneDrive\Dokumenty\4th semester\Artificial Intelligence\Lab\Project\Accident Detection.v1i.tensorflow\test\test_img_no_100_jpg.rf.084364eb115b3c7fc246526f24806e63.jpg"
result = CLIENT.infer(image_path, model_id="accident-detection-xymwp/1")
```

```python
image = Image.open(image_path)
```

```python
fig, ax = plt.subplots(1)
ax.imshow(image)

detections = result["predictions"]

for detection in detections:
    x, y, width, height = detection['x'], detection['y'], detection['width'], detection['height']
    left = x - width / 2
    top = y - height / 2
    label = detection['class']
    confidence = detection['confidence']

    rect = patches.Rectangle((left, top), width, height, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    plt.text(left, top - 10, f'{label} {confidence:.2f}', color='red', fontsize=8, weight='bold')

plt.show()
```

```python
def model_predict(image_path):
    result = CLIENT.infer(image_path, model_id="accident-detection-xymwp/1")
    image = Image.open(image_path)
    fig, ax = plt.subplots(1)
    ax.imshow(image)

    detections = result["predictions"]

    for detection in detections:
        x, y, width, height = detection['x'], detection['y'], detection['width'], detection['height']
        left = x - width / 2
        top = y - height / 2
        label = detection['class']
        confidence = detection['confidence']

        rect = patches.Rectangle((left, top), width, height, linewidth=1, edgecolor='r', facecolor='none')
        ax.add_patch(rect)

        plt.text(left, top - 10, f'{label} {confidence:.2f}', color='red', fontsize=8, weight='bold')

    plt.show()
```
