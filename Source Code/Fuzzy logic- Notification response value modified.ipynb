```python
import skfuzzy as fuzz
data = combined_df
```

```python
criticality = np.arange(0, 400, 1)
emergency_score = np.arange(0, 2, 1)
risk_score = np.arange(0, 1500, 1)
severity = np.arange(1, 5, 1)

crit_lo = fuzz.trimf(criticality, [0, 0, 100])
crit_md = fuzz.trimf(criticality, [0, 100, 200])
crit_hi = fuzz.trimf(criticality, [100, 200, 400])

emer_lo = fuzz.trimf(emergency_score, [0, 0, 1])
emer_hi = fuzz.trimf(emergency_score, [0, 1, 1])

risk_lo = fuzz.trimf(risk_score, [0, 0, 300])
risk_md = fuzz.trimf(risk_score, [200, 600, 1000])
risk_hi = fuzz.trimf(risk_score, [800, 1000, 1500])

sever_lo = fuzz.trimf(severity, [1, 1, 2])
sever_md = fuzz.trimf(severity, [1, 2, 3])
sever_hi = fuzz.trimf(severity, [2, 3, 4])
sever_vhi = fuzz.trimf(severity, [3, 4, 4])
```

```python
weights = {
    'criticality': 0.1,
    'emergency_score': 0.1,
    'risk_score': 0.4,
    'Severity': 0.4
}
```

```python
def calculate_fuzzy_weighted_score(row):
    crit_level_lo = fuzz.interp_membership(criticality, crit_lo, row['criticality'])
    crit_level_md = fuzz.interp_membership(criticality, crit_md, row['criticality'])
    crit_level_hi = fuzz.interp_membership(criticality, crit_hi, row['criticality'])
    emer_level_lo = fuzz.interp_membership(emergency_score, emer_lo, row['emergency_score'])
    emer_level_hi = fuzz.interp_membership(emergency_score, emer_hi, row['emergency_score'])
    risk_level_lo = fuzz.interp_membership(risk_score, risk_lo, row['risk_score'])
    risk_level_md = fuzz.interp_membership(risk_score, risk_md, row['risk_score'])
    risk_level_hi = fuzz.interp_membership(risk_score, risk_hi, row['risk_score'])
    sever_level_lo = fuzz.interp_membership(severity, sever_lo, row['Severity'])
    sever_level_md = fuzz.interp_membership(severity, sever_md, row['Severity'])
    sever_level_hi = fuzz.interp_membership(severity, sever_hi, row['Severity'])
    sever_level_vhi = fuzz.interp_membership(severity, sever_vhi, row['Severity'])
    weighted_fuzzy_sum = (
        weights['criticality'] * max(crit_level_lo, crit_level_md, crit_level_hi) +
        weights['emergency_score'] * max(emer_level_lo, emer_level_hi) +
        weights['risk_score'] * (risk_level_lo + 2 * risk_level_md + 3 * risk_level_hi) / 6 +
        weights['Severity'] * (sever_level_lo + 2 * sever_level_md + 3 * sever_level_hi + 4 * sever_level_vhi) / 10
    )
    return weighted_fuzzy_sum
```

```python
data['fuzzy_score'] = data.apply(calculate_fuzzy_weighted_score, axis=1)

min_score = data['fuzzy_score'].min()
max_score = data['fuzzy_score'].max()
data['normalized_fuzzy_score'] = (data['fuzzy_score'] - min_score) / (max_score - min_score)
```

## Emergency Services Allocation Fuzzy Rules

**Fire Fighters:**
Severity >= 3
normalized_fuzzy_score > 0.5
criticality > 100

**Ambulance with EMTs:**
Severity >= 2
normalized_fuzzy_score > 0.3
risk_score > 75

**Police Officers:**
Severity >= 2
normalized_fuzzy_score > 0.2
risk_score > 50

**Tow Truck Operators:**
Severity >= 1
normalized_fuzzy_score > 0.1
emergency_score > 50

**Traffic Control Authorities:**
Always required

```python
def determine_services(row):
    services_needed = ["Traffic Control Authorities"]  # Always required

    if row['Severity'] >= 3 or row['normalized_fuzzy_score'] > 0.5 or row['criticality'] > 100:
        services_needed.append("Fire Fighters")

    if row['Severity'] >= 2 or row['normalized_fuzzy_score'] > 0.3 or row['risk_score'] > 75:
        services_needed.append("Ambulance with EMTs")

    if row['Severity'] >= 2 or row['normalized_fuzzy_score'] > 0.2 or row['risk_score'] > 50:
        services_needed.append("Police Officers")

    if row['Severity'] >= 1 or row['normalized_fuzzy_score'] > 0.1 or row['emergency_score'] > 50:
        services_needed.append("Tow Truck Operators")

    return list(set(services_needed))  # Remove any duplicates

# Apply the function to the DataFrame
data["Services_Needed"] = data.apply(determine_services, axis=1)

# Display the DataFrame with horizontal scrolling
html_table = data.to_html(classes="dataframe")
html_table_with_style = f"<div class='dataframe-container'>{html_table}</div>"
display(HTML(html_table_with_style))
```
