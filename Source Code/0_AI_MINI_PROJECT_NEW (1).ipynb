### Latitude Longitude Generator

```python
import json
from pyproj import Proj, transform

def add_lat_lon(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)

    for entry in data:
        if 'pos' in entry:  # Check if 'pos' attribute exists
            pos = entry['pos']
            lat, lon = convert_to_lat_lon(pos)
            entry['latitude'] = lat
            entry['longitude'] = lon

    with open(json_file, 'w') as file:
        json.dump(data, file, indent=4)

def convert_to_lat_lon(pos):
    # Assuming pos is in meters, you may need to adjust the conversion factor based on your coordinate system
    x, y, _ = pos  # Ignore the z-coordinate
    # You need to specify your coordinate system. Assuming it's a simple Mercator projection here.
    in_proj = Proj(init='epsg:3857')
    out_proj = Proj(init='epsg:4326')  # WGS84 coordinate system (latitude and longitude)
    lon, lat = transform(in_proj, out_proj, x, y)
    return lat, lon

add_lat_lon("C:/Users/DELL/Downloads/can_bus/can_bus/scene-0001_pose.json")
```

### Engine RPM Calculator

```python
import json

def calculate_mean_eff_torque_stats(json_file):
    with open(json_file, 'r') as file:
        data = json.load(file)

    # Initialize variables to store statistics
    max_value = float('-inf')
    min_value = float('inf')
    total_value = 0
    count = 0

    # Iterate through each entry in the JSON data
    for entry in data:
        mean_eff_torque = entry.get('meanEffTorque', None)
        if mean_eff_torque is not None:
            max_value = max(max_value, mean_eff_torque)
            min_value = min(min_value, mean_eff_torque)
            total_value += mean_eff_torque
            count += 1

    # Calculate average value
    average_value = total_value / count if count > 0 else None

    print("Max_Torque: ",max_value, "Min_Torque: ", min_value,"Average_Torque: ",average_value)
    return max_value, min_value, average_value

def estimate_engine_rpm(json_file,max_toque,min_torque):
    with open(json_file, 'r') as file:
        data = json.load(file)

    # Iterate through each entry in the JSON data
    for entry in data:
        mean_eff_torque = entry.get('meanEffTorque', None)

        # Assuming linear relationship between torque and RPM
        torque_constant = 40  # Constant to convert torque to RPM (simplified)

        # Adjust torque value to ensure it falls within a reasonable range for RPM calculation
        adjusted_torque = max(min(mean_eff_torque, max_torque), min_torque)

        if adjusted_torque < 0 :
            torque_constant = 8

        # Estimate the change in RPM based on the torque
        rpm_change = adjusted_torque * torque_constant

        # Assuming initial RPM value
        initial_rpm = 1000  # Example initial RPM value

        # Update the current RPM value
        current_rpm = initial_rpm + rpm_change

        entry['engine_rpm'] = current_rpm

    # Write updated data back to the file
    with open(json_file, 'w') as file:
        json.dump(data, file, indent=4)

# Replace 'file_path' with the path to your JSON file
file_path = 'C:/Users/DELL/Downloads/can_bus/can_bus/scene-0001_zoe_veh_info.json'
max_torque, min_torque, average_torque = calculate_mean_eff_torque_stats(file_path)
estimate_engine_rpm(file_path,max_torque,min_torque)
```

### Function to filter individual sensor data for each scene

```python
import pandas as pd
import json

def filter_attributes(file_path):
    # Load the JSON data
    with open(file_path, 'r') as f:
        data = json.load(f)

    # List of necessary attributes
    necessary_attributes = ["utime", "linear_accel", "rotation_rate", 'brake_sensor', 'steering_sensor', 'throttle_sensor', "value","brake", "meanEffTorque", "steering", "steering_speed", "throttle", "vehicle_speed", "FL_wheel_speed", "FR_wheel_speed","RL_wheel_speed","RR_wheel_speed",""]

    # Filter out unnecessary attributes and handle missing attributes
    filtered_data = []
    for entry in data:
        filtered_entry = {}
        for attr in necessary_attributes:
            if attr in entry:
                if attr == 'utime':
                    # Crop utime attribute value to first 10 digits
                    filtered_entry[attr] = str(entry[attr])[:10]
                else:
                    filtered_entry[attr] = entry[attr]

        filtered_data.append(filtered_entry)

    # Create DataFrame
    df = pd.DataFrame(filtered_data)

    # Save DataFrame to a new JSON file
    filtered_file_path=file_path.replace('can_bus/can_bus','can_bus/lan_bus')
    filtered_file_path = filtered_file_path.replace('.json', '_filtered.json')
    df.to_json(filtered_file_path, orient='records')

    print(f"Filtered dataset saved to: {filtered_file_path}")

```
